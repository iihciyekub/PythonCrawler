- 使用私有属性,类方法访问类静态字段(保证数据的准确性)

```python
class typea():
    __download_num = 0

    @classmethod
    def getnum(cls):
        print(cls.__download_num)

    def __init__(self, name):
        self.name = name

    def open(self):
        print(f"打开{self.name}")
        print('去掉广告')

    def search(self, keyword):
        print(f'输入关键词{keyword},开始检索')

    @classmethod
    def get_result(cls):
        print('收集所有资讯')
        print('下载资讯')
        cls.__download_num += 1

    def letdo(self, keyword):
        self.open()
        self.search(keyword)
        self.get_result()


class typeb(typea):
    __share_num = 0

    @classmethod
    def getnum(cls):
        print(cls.__share_num)

    @classmethod
    def get_result(cls):
        print('找到有用的资讯并分享')
        cls.__share_num += 1


a = typea("百度")
b = typeb("谷歌")
a.letdo("pythonn")

print("--------------")
b.letdo("pythonn")
b.letdo("pythonn")
b.letdo("pythonn")
print("--------------")
a.download_num = 9999
b.share_num = 9999
a.getnum()
b.getnum()
```

- 使用私有属性,以及类的特性锁定只读属性

```python
class typea():
    __download_num = 0  #私有属性

    @property  #设置只读属性
    def download_num(self):
        return self.__class__.__download_num

    def __init__(self, name):
        self.name = name

    def open(self):
        print(f"打开{self.name}")
        print('去掉广告')

    def search(self, keyword):
        print(f'输入关键词{keyword},开始检索')

    @classmethod
    def get_result(cls):
        print('收集所有资讯')
        print('下载资讯')
        cls.__download_num += 1

    def letdo(self, keyword):
        self.open()
        self.search(keyword)
        self.get_result()


class typeb(typea):
    __share_num = 0  #私有属性

    @property  #设置只读属性
    def share_num(self):
        return self.__class__.__share_num

    @classmethod
    def get_result(cls):
        print('找到有用的资讯并分享')
        cls.__share_num += 1


a = typea("百度")
b = typeb("谷歌")
a.letdo("pythonn")
a.letdo("pythonn")
print("--------------")
b.letdo("pythonn")
b.letdo("pythonn")
b.letdo("pythonn")
print("--------------")

print(a.download_num)
print(b.share_num)
```
